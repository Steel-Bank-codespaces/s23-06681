# Print lines in a file

We open a file, and then iterate for each line in the file. Here is an example.

The variable `line` contains a string.



```python
import os
with open(os.path.expanduser('~/s23-06681/lectures/01-numpy/p-t.dat')) as f:
    for line in f:
        print(line)

```

```python
type(line)

```

How do we get stuff out of that string? We can split it into a list of strings to start.



```python
line.split()

```

That is closer, but we want numbers. We have to convert those strings to numbers. You can get an integer with `int`, and a float with `float`. Here is an example.



```python
int('40'), float('-1.251')

```

That is still not quite all you need. Let's turn the first two numbers to ints and the last four numbers to floats in a list comprehension. 



```python
fields = line.split()
[int(x) for x in fields[0:2]] + [float(x) for x in fields[2:]]

```
